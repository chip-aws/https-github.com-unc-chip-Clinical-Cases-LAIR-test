## What is Data Visualization?

Data visualization is the presentation of data in a pictorial or graphical format. It enables decision makers to see analytics presented visually, so they can grasp difficult concepts or identify trends and patterns. With the rise of big data upon us, we need efficient tools to help us interpret the rapidly growing data. As data analytics reveals us the fact behind the data, visualization gives us as human beings an intuition visually on how the data is actually related to the fact.

Data visualization is not exclusive for the digital era, it has been existing for a long time. The earliest and most obvious examples of dataviz projects are maps. Then you have the pie chart which first showed up in the early 19th century.  A few decades later, Charles Joseph Minard used statistical graphs in order to map Napoleon Bonapartes Russian campaign of 1812, combining multiple metrics: the number of troops, temperature, distance, directions, and more. ## What We Can Do with Data Visualization? 

### Comparing Data

For comparing data side by side, Bar Chart is a good choice. The following example shows the top 10 products by revenue in two regions. From the chart, we can easily tell the difference among the products as well as between the two regions.


Besides Bar Charts, we also have [Stack Charts](https://www.anychart.com/chartopedia/chart-type/percent-stacked-area-chart/) and [Radar Charts](https://www.anychart.com/chartopedia/chart-type/radar-chart/) that can help us with comparing data. Explore the links for interactive examples.

### Exploring Composition of Data

For exploring the composition of data, Pie Chart and [Tree Map](https://www.anychart.com/chartopedia/chart-type/treemap/) are good choices. The following picture shows the the operating system market share of consoles in a pie chart. Explore the link of Tree Map for an interactive example. 



### Tracking the Trends of Data

For tracking the trends, [line charts](https://www.anychart.com/chartopedia/chart-type/line-chart/) are frequently used. The following example shows a trend or pattern of the server CPU load by days in a week. By adding more than one lines, we can see a comparison between the trends of different objects.


By specifying the area between or under the lines, we can have a better understanding of the cumulative data. This is what we call [Area Chart](https://www.anychart.com/chartopedia/chart-type/area-chart/). 

### Analyzing Data Distribution

To analyze data distribution and the relationship between datasets. [Scatter Charts](https://www.anychart.com/chartopedia/chart-type/dot-chart/) are very helpful. The following Scatter Chart shows the relationship between the weight and height parameters. It is not absolute, but from the chart we can see a pretty strong positive relationship between these two parameters.


Besides, to show the relationship and distribution, [Box Charts](https://www.anychart.com/chartopedia/chart-type/box-chart/), [Bubble Charts](https://www.anychart.com/chartopedia/chart-type/bubble-chart/), [Heat Map Charts](https://www.anychart.com/chartopedia/chart-type/heatmap/) and [Polar Charts](https://www.anychart.com/products/anychart/gallery/Polar_Charts/) are also good choices.

### Making Sense of Geographical Data

There are a lot of information you may want to visulize geographically such as the number of cases of COVID-19, the support rate of president election, etc. The following example shows a Choropleth Map of the United States presidential election in 2008. 


Depends on the color being used, [Choropleth Map](https://www.anychart.com/chartopedia/chart-type/choropleth-map/) can show contrast between different categories or different degrees of the same category (the intensity of population and the tempereture of different areas). Besides using colors, we can also use dots to mark the regions as in [Dot Maps](https://www.anychart.com/chartopedia/chart-type/dot-map/), indicate volumns using [Bubble Maps](https://www.anychart.com/chartopedia/chart-type/bubble-map/), show connections as in [Connector Maps](https://www.anychart.com/chartopedia/chart-type/connector-map/), and show movement as in [Flow Maps](https://www.anychart.com/chartopedia/chart-type/flow-map/).

## Python Data Visualization Packages

### Data visualizations via LEATHER

As stated in its document, [Leather](https://pypi.org/project/leather/) is the Python charting library for those who need charts now and dont care if theyre perfect. It is a easy-to-use, popular data visualization library for Python. First of all, let's import the library.

```python
import leather
```

#### Scatter Chart



To show a quick example of visulization, let's firstly generate a few data points.

```python
data1 = ([1.5, 2], [2, 3], [4.5, 6], [7.5, 4])
```

By running the following block of code, you'll be able to draw the data points on a chart.

```python
## Visualizing simple dots
chart = leather.Chart('Simple pairs of x-y')
chart.add_dots(data1)
chart.to_svg()
```

#### Line Chart

To link these point into a line and show the trends, we just need to use the function add_line().

```python
## Visualizing lines
chart = leather.Chart('Visualizing Lines')
chart.add_line(data1)
chart.to_svg()
```

Good job! Let's add another line to the chart.

```python
## Have multiple lines in one plot
data2 = ([3, 2], [4, 1.5], [4.5, 3], [7, 4])

chart = leather.Chart('Visualizing Lines')
chart.add_line(data1)
chart.add_line(data2)
chart.to_svg()
```

#### Bar Chart

Let's draw a Bar Chart this time. First of all, let's initialize the data.

```python
data3 = [('1-3', 13), ('4-9', 5),('10-14', 15), ('15-21', 3)]
```

```python
## Visualize data in bar charts
chart = leather.Chart('Bar charts')
chart.add_columns(data3)
chart.to_svg()
```

### Visualizing via matplotlib

Different from Leather which was designed to be easy-to-use, basic, but non-perfect, Matplotlib produces publication-quality figures in a variety of hardcopy formats and interactive environments across platforms with complex tuners and parameters. In this module, we'll show several simple examples using matplotlib. But keep in mind, matplotlib is much more powerful than this! First of all, let's import the libraries we need.

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Basic Visualization in Dots and Lines. 

Before plotting, we need to create some data points. This time we use two arrays each contains numbers from 0 to 9.

```python
## Create ndarrays
x_arry = np.arange(10)
y_arry = np.arange(10)
```

Then we use the plot function to compose the chart. Feel free to substitute "o--" with "--" and "o" to see the difference.

```python
## Visulize the arrays using Matplotlib
plt.plot(x_arry, y_arry, 'o--')
plt.show()
```

#### Visualizing a Function

matplotlib allows you to visualize a function and see how the dependent parameter(y) has been changed according to the independent parameter(x). For an example, let's try to visualize the function y = f(x) = 2x^2 + 4 . Let's give the x a range so that our visualization becomes more practical. We initialize the x with the odd number from 3 to 11.

```python
x = [3, 5, 7, 9 ,11]
```

Then we can try to plot the chart showing the relationship between x and y.

```python
plt.plot(x, [2*i**2 + 4 for i in x])
```

Looks like we've already have a function visualization! How about we refine it a little bit by adding some features? Let's add a grid, x and y labels, and a title to make it easier to read.

```python
## How to visualize a function
## Ex, y = f(x) = 2x^2 + 4
## x = [3, 5, 7, 9 ,11]

plt.plot(x, [2*i**2 + 4 for i in x]) # plot the function
plt.grid(True) ## add grid
plt.axis([2, 12, 20, 260]) # define the range of the grid
plt.xlabel('x = odd number from 3 to 11') # adding the x-label
plt.ylabel('y = f(x) = 2x^2 + 4') # adding the y-label
plt.title('Simple function plot') # adding the title
plt.show()
```

#### Visualizing a 3D Chart

matplotlib allows visualization in 3D. Here we'll need to import some new packages.

```python
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import axes3d
```

Firstly, we need to generate an empty figure. For 3D visualizations we need to enable 3D projections with the parameter and argument pair projection='3d'. Then, we'll generate some data for the X, Y, and Z axis as we are drawing a 3D figure. Here we use the function get_test_data() to get some test data for our demonstration.

```python
fig = plt.figure() # generating an empty figure
ax = fig.gca(projection='3d') # specfying the 3d parameter
X, Y, Z = axes3d.get_test_data(delta=0.1) # assigning the X, Y, and Z axis with some test data
ax.plot_wireframe(X, Y, Z) # buidling the frame
plt.show() # showing the plot
```

#### Network visualizations via networkx and matplotlib

A network is always used to express connection. For example, the transportation among cities can be represented as a network where each city is represented as a node and the transportations are represented as edges. Another example is to show the relationship among a group of people, in which the people are nodes while the relationships are edges. Let's see how we can visualize a network using networkx. 

```python
import networkx as nx
```

A network is represented in a Graph. Firstly, let's initialize empty a Graph.

```python
## Creating an empty graph
G = nx.Graph()
```

Then we add nodes to the Graph. In practice, these nodes can be used to represent objects such as cities and people.

```python
G.add_nodes_from(['a', 'b', 'c', 'd', 'e', 'f']) # adding nodes
nx.draw(G, with_labels=True) # draw the network
plt.show() # visualize the network using matplotlib
```

After adding nodes, we'll link them together by adding edges. As we don't specify the distance between the nodes, the network might look random.

```python
G.add_edges_from([('a','e'), ('a','c'), ('c','d'), ('e','f'), ('f','a'), ('f','b')])
nx.draw(G, with_labels=True)
plt.show()
```

## Covid-19 demo

Covid is a Python package to get information regarding the novel corona virus provided by Johns Hopkins university and worldometers.info. We'll be using the cleaned data provided in this package for our demonstration. First of all, let's import the packages we're going to use, covid and pandas.

```python
from covid import Covid
import pandas as pd

## fetch the data
## by default data source is "john_hopkins"
covid = Covid() ## or covid = Covid(source="john_hopkins")

## to get data from worldometers.info
covid = Covid(source="worldometers")

## get all data
covid.get_data()
```

Next, we'll import the data to pandas dataframe for further use. You'll learn more about pandas in other online modules.

```python
## Import the data to pandas dataframe
df = pd.DataFrame(covid.get_data())
print(df)
```

In our demo, we'd like to visualize the covid situation (confirmed, active, deaths cases etc.) according to regions. Let's focus on the first ten regions that have the most confirmed case number. To do this, We need to sort our data according to the confirmed cases.

```python
## Organize data for visualization
sorted_df = df.sort_values(by=['confirmed'], ascending=False)
sorted_df
```

Then, we use the following code to exclude the heading from the table and extract the top 10 records.

```python
## Excluding the heading from the table and extract top 10 records
example = sorted_df.head(10)
example
```

This data contains the cumulative data for all the continents and the world, too. We can exclude the data for the world and the continents as follows:

```python
excluded = sorted_df[~sorted_df.country.isin(['Europe', 'South America', 'Asia', 'World', 'North America', 'Africa'])]
excluded
```

Now we extract the top 10 countries that have the most confirmed cases.

```python
top10 = excluded.head(10)
top10
```

To draw a 2D chart, we need the values for x and y axis. This time, let's take a look at the comparison of confirmed cases among the top 10 regions using a bar chart.

```python
x = top10.country
y1 = top10.confirmed
plt.bar(x,y1) # drawing the bar chart for variable x(countries) and y(confirmed cases)
plt.xticks(rotation=90) # to rotate the x labels 90 degrees.
plt.show() # show the chart
```

Bar chart is good for comparing numbers, besides, we can also use a line chart which is also good for displaying the difference among the countries.

```python
plt.plot(x,y1) # drawing the bar chart for variable x(countries) and y(confirmed cases)
plt.xticks(rotation=90) # to rotate the x labels 90 degrees.
plt.show() # show the chart
```

Now let's try some other variables. For example, the active cases.

```python
# Define the variable
y2 = top10.active

## Visualize the data
plt.bar(x, y2)
plt.xticks(rotation=90)
plt.show()
```
